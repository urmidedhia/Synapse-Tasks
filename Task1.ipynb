```python
import cv2
```

```python
img = cv2.imread("SynapseTask1.png")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
(T, thresh) = cv2.threshold(img2, 0, 255, cv2.THRESH_BINARY_INV)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
largest = [0, []]
for c in contours:
    contour_area = cv2.contourArea(c)
    if contour_area > largest[0]:
        largest = [contour_area, c]
```

```python
cv2.drawContours(img, largest[1], -1, (0, 255, 0), 10)
```

```python
cv2.imshow("All contours", img)
```
